<h3> Replacing Outliers with Median </h3>

```python
import numpy as np

def outlier_detect(df):
    # Select only numerical columns
    numerical_columns = df.select_dtypes(include='number').columns
    
    for column in numerical_columns:
        # Calculate 25th percentile and 75th percentile
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # Detect outliers
        outliers_check = (df[column] < lower_bound) | (df[column] > upper_bound)
        print (df.loc[outliers_check, column])
        # Replace outliers with the median of the column
        median_value = df[column].median()
        df.loc[outliers_check, column] = median_value
    
    return df

# Usage
df_new=outlier_detect(df)
```

<h3> How do we get to know if we have removed the outliers? </h3>
<p>25% (Q1) and 75% (Q3) percentiles: These are the key values for detecting outliers.</p>
<p>Min and Max Compare these with the outlier bounds you calculated. If your lower_bound and upper_bound exclude the min or max values, then those are outliers. If the min/max values are within those bounds, you won’t have any outliers.</p>

```python
from zlib import crc32
import numpy as np
```

```python
def test_set_check(identifier, test_ratio=0.25):
    total_size = 2**32
    hex_repr = crc32(np.int64(identifier)) & 0xffffffff
    in_test = hex_repr < (test_ratio * total_size)
    return in_test
```

```python
[test_set_check(i) for i in range(10)]
```

```python
def split_train_test_by_id(data, test_ratio):
    ids = data['index']
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
```

<h3> Normal Form </h3>
<p> As our happiness dataset contains multiple input features and 1 target variable or label (Life Ladder), we will first compute the values using Normal Equation, in this we will first compute the best weights i.e. theta_best first in one go and not multiple iterations and then use the best weights to compute the predications using dot product with the input features. </p>

```python
import numpy as np


X = happiness_feature_prep.values  
y = happiness_label_prep.values     

# Adding bias 
X_b = np.c_[np.ones((X.shape[0], 1)), X] 

#Normal Equation
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

predictions = X_b.dot(theta_best)

print("Weights (theta):", theta_best)
print("Prediction:", predictions)

```

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as mp
X_SGD = happiness_feature_prep.values  # Features
y_SGD = happiness_label_prep.values     # Target/Label

## We are again spliting the data into Train-Test Split in order to check for Training and 
## Validation Loss (Overfitting and Underfitting) and also for hyperparam tuning!

X_train_SGD, X_val_SGD, y_train_SGD, y_val_SGD = train_test_split(X_SGD, y_SGD, test_size=0.25, random_state=42)

# Add a bias term (intercept) to the feature matrix
X_train_b_SGD = np.c_[np.ones((X_train_SGD.shape[0], 1)), X_train_SGD]  
X_val_b_SGD = np.c_[np.ones((X_val_SGD.shape[0], 1)), X_val_SGD]        


theta = np.random.randn(X_train_b_SGD.shape[1])  # For #1 Run, randomly intializing the weights


learning_rate = 0.001
num_epochs = 150
batch_size =16


training_losses = []
validation_losses = []

# Step 4: Stochastic Gradient Descent
for epoch in range(num_epochs):
    # Shuffle the training data
    indices = np.random.permutation(len(X_train_b_SGD))
    X_train_b_shuffled = X_train_b_SGD[indices]
    y_train_shuffled = y_train_SGD[indices]

    # Mini-batch gradient descent
    for i in range(0, len(X_train_b_shuffled), batch_size):
        X_batch = X_train_b_shuffled[i:i + batch_size]
        y_batch = y_train_shuffled[i:i + batch_size]

        
        y_pred = X_batch @ theta

        
        gradients = -2 / len(X_batch) * X_batch.T @ (y_batch - y_pred)

        # Updating weights
        theta -= learning_rate * gradients

    # Calculate Training RMSE after each epoch
    y_train_pred_SGD = X_train_b_SGD @ theta
    train_rmse_SGD = np.sqrt(np.mean((y_train_SGD - y_train_pred_SGD) ** 2))
    training_losses.append(train_rmse_SGD)

    # Calculate validation RMSE
    y_val_pred_SGD = X_val_b_SGD @ theta
    val_rmse_SGD = np.sqrt(np.mean((y_val_SGD - y_val_pred_SGD) ** 2))
    validation_losses.append(val_rmse_SGD)

    # Print the training and validation RMSE for this epoch
    print(f'Epoch {epoch + 1}/{num_epochs}, Train RMSE: {train_rmse_SGD:.4f}, Val RMSE: {val_rmse_SGD:.4f}, Diff: {train_rmse_SGD-val_rmse_SGD:4f}')

# Step 5: Plot the Losses
mp.plot(training_losses, label='Train RMSE', color='blue', marker='o')
mp.plot(validation_losses, label='Val RMSE', color='red', marker='x')
mp.xlabel('Epochs')
mp.ylabel('RMSE')
mp.title('Training and Validation RMSE')
mp.legend()
mp.show()
```

<h3> Ridge Regression </h3
<p> In Ridge Regression we'll be considering different values of Alpha (Regularization Values) and their impact on Training and Validation RMSE </p>

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import numpy as np

X=happiness_feature_prep.values
y=happiness_label_prep.values

X_train_ridge, X_val_ridge, y_train_ridge, y_val_ridge = train_test_split(X, y, test_size=0.25, random_state=42)
# Different values for alpha
alphas = [0.001,0.01 ,0.1, 0.2, 0.5, 1]

training_losses_ridge = []
validation_losses_ridge = []

for alpha in alphas:
    # Initialize Ridge Regression model with current alpha
    ridge_reg = Ridge(alpha=alpha)
    
    # Fit the model on the training data
    ridge_reg.fit(X_train_ridge, y_train_ridge)
    
    # Make predictions on training and validation sets
    y_train_pred_ridge = ridge_reg.predict(X_train_ridge)
    y_val_pred_ridge = ridge_reg.predict(X_val_ridge)
    
    # Calculate RMSE for training and validation sets
    train_rmse_ridge = np.sqrt(mean_squared_error(y_train_ridge, y_train_pred_ridge))
    val_rmse_ridge = np.sqrt(mean_squared_error(y_val_ridge, y_val_pred_ridge))
    
    training_losses_ridge.append(train_rmse_ridge)
    validation_losses_ridge.append(val_rmse_ridge)
    
    print(f'Alpha: {alpha}')
    print(f'Ridge Training RMSE: {train_rmse_ridge:.4f}')
    print(f'Ridge Validation RMSE: {val_rmse_ridge:.4f}')

# Plotting the RMSE vs Alpha
plt.figure(figsize=(8, 6))
plt.plot(alphas, training_losses_ridge, label='Training RMSE', marker='o')
plt.plot(alphas, validation_losses_ridge, label='Validation RMSE', marker='x')
plt.xlabel('Alpha')
plt.ylabel('RMSE')
plt.title('Ridge Regression: RMSE vs Alpha')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as mp

# Features and Labels
X_SGD = happiness_feature_prep.values  # Features
y_SGD = happiness_label_prep.values    # Target/Label

# Train-Test Split
X_train_SGD, X_val_SGD, y_train_SGD, y_val_SGD = train_test_split(X_SGD, y_SGD, test_size=0.25, random_state=42)

# Add a bias term (intercept) to the feature matrix
X_train_b_SGD = np.c_[np.ones((X_train_SGD.shape[0], 1)), X_train_SGD]  
X_val_b_SGD = np.c_[np.ones((X_val_SGD.shape[0], 1)), X_val_SGD]        

theta = np.random.randn(X_train_b_SGD.shape[1])


learning_rate = 0.01
num_epochs = 100
batch_size = 32
alpha = 0.001  # Regularization strength (Ridge regularization)

# To store RMSE during training
training_losses = []
validation_losses = []

# Stochastic Gradient Descent with Ridge (L2 Regularization)
for epoch in range(num_epochs):
    # Shuffle the training data
    indices = np.random.permutation(len(X_train_b_SGD))
    X_train_b_shuffled = X_train_b_SGD[indices]
    y_train_shuffled = y_train_SGD[indices]
    #print (y_train_shuffled)
    #print(X_train_b_shuffled)

    # Mini-batch gradient descent
    for i in range(0, len(X_train_b_shuffled), batch_size):
        X_batch = X_train_b_shuffled[i:i + batch_size]
        y_batch = y_train_shuffled[i:i + batch_size]

        # Predict the output
        y_pred = X_batch @ theta

        # Compute the gradients with Ridge regularization (L2)
        gradients = -2 / len(X_batch) * X_batch.T @ (y_batch - y_pred) + 2 * alpha * theta  # L2 penalty term

        # Update the weights
        theta -= learning_rate * gradients

    # Calculate TRMSE
    y_train_pred_SGD = X_train_b_SGD @ theta
    train_rmse_SGD = np.sqrt(np.mean((y_train_SGD - y_train_pred_SGD) ** 2))
    training_losses.append(train_rmse_SGD)

    # Calculate VRMSE
    y_val_pred_SGD = X_val_b_SGD @ theta
    val_rmse_SGD = np.sqrt(np.mean((y_val_SGD - y_val_pred_SGD) ** 2))
    validation_losses.append(val_rmse_SGD)

    # Printing the TRMSE and VRMSE for each epoch
    print(f'Epoch {epoch + 1}/{num_epochs}, Train RMSE: {train_rmse_SGD:.4f}, Val RMSE: {val_rmse_SGD:.4f}, Diff: {train_rmse_SGD-val_rmse_SGD:.4f}')

# Step 5: Plotting
mp.plot(training_losses, label='Train RMSE', color='blue', marker='o')
mp.plot(validation_losses, label='Val RMSE', color='red', marker='x')
mp.xlabel('Epochs')
mp.ylabel('RMSE')
mp.title('Training and Validation RMSE (Ridge SGD)')
mp.legend()
mp.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
import matplotlib.pyplot as mp

# Features and Labels
X_SGD = happiness_feature_prep.values  # Features
y_SGD = happiness_label_prep.values    # Target/Label

# Define degree for polynomial regression
degree = 2 # You can adjust this for different polynomial degrees

# Create polynomial features
poly = PolynomialFeatures(degree=degree)
X_poly_SGD = poly.fit_transform(X_SGD)

# Train-Test Split for polynomial features
X_train_SGD, X_val_SGD, y_train_SGD, y_val_SGD = train_test_split(X_poly_SGD, y_SGD, test_size=0.25, random_state=42)

# Add a bias term (intercept) to the feature matrix
X_train_b_SGD = np.c_[np.ones((X_train_SGD.shape[0], 1)), X_train_SGD]  
X_val_b_SGD = np.c_[np.ones((X_val_SGD.shape[0], 1)), X_val_SGD]        

# Initialize theta randomly (weights)
theta = np.random.randn(X_train_b_SGD.shape[1])

# Hyperparameters
learning_rate = 0.01
num_epochs = 100
batch_size = 32

# To store RMSE during training
training_losses = []
validation_losses = []

# Stochastic Gradient Descent without regularization
for epoch in range(num_epochs):
    # Shuffle the training data
    indices = np.random.permutation(len(X_train_b_SGD))
    X_train_b_shuffled = X_train_b_SGD[indices]
    y_train_shuffled = y_train_SGD[indices]

    # Mini-batch gradient descent
    for i in range(0, len(X_train_b_shuffled), batch_size):
        X_batch = X_train_b_shuffled[i:i + batch_size]
        y_batch = y_train_shuffled[i:i + batch_size]

        # Predict the output
        y_pred = X_batch @ theta

        # Compute the gradients (No Regularization)
        gradients = -2 / len(X_batch) * X_batch.T @ (y_batch - y_pred)

        # Update the weights
        theta -= learning_rate * gradients

    # Calculate Training RMSE (TRMSE)
    y_train_pred_SGD = X_train_b_SGD @ theta
    train_rmse_SGD = np.sqrt(np.mean((y_train_SGD - y_train_pred_SGD) ** 2))
    training_losses.append(train_rmse_SGD)

    # Calculate Validation RMSE (VRMSE)
    y_val_pred_SGD = X_val_b_SGD @ theta
    val_rmse_SGD = np.sqrt(np.mean((y_val_SGD - y_val_pred_SGD) ** 2))
    validation_losses.append(val_rmse_SGD)

    # Print the training and validation RMSE for each epoch
    print(f'Epoch {epoch + 1}/{num_epochs}, Train RMSE: {train_rmse_SGD:.4f}, Val RMSE: {val_rmse_SGD:.4f}, Diff: {train_rmse_SGD-val_rmse_SGD:.4f}')

# Plotting the training and validation losses (RMSE)
mp.plot(training_losses, label='Train RMSE', color='blue', marker='o')
mp.plot(validation_losses, label='Val RMSE', color='red', marker='x')
mp.xlabel('Epochs')
mp.ylabel('RMSE')
mp.title('Training and Validation RMSE (Polynomial Regression with SGD)')
mp.legend()
mp.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as mp

# Features and Labels
X_SGD = happiness_feature_prep.values  # Features
y_SGD = happiness_label_prep.values    # Target/Label

# Define degree of polynomial
degree = 2  # Adjust the degree as needed

# Transform features into polynomial features
poly = PolynomialFeatures(degree=degree)
X_poly = poly.fit_transform(X_SGD)

# Train-Test Split
X_train_SGD, X_val_SGD, y_train_SGD, y_val_SGD = train_test_split(X_poly, y_SGD, test_size=0.20, random_state=42)

# Add a bias term (intercept) to the feature matrix
X_train_b_SGD = np.c_[np.ones((X_train_SGD.shape[0], 1)), X_train_SGD]  
X_val_b_SGD = np.c_[np.ones((X_val_SGD.shape[0], 1)), X_val_SGD]        

# Initialize weights randomly
theta = np.random.randn(X_train_b_SGD.shape[1])

# Hyperparameters
learning_rate = 0.01
num_epochs = 100
batch_size = 32
alpha = 0.001 # Regularization strength
l1_ratio = 0.001  # Elastic Net mixing ratio (0: Ridge, 1: Lasso, in between: Elastic Net)

# To store RMSE during training
training_losses = []
validation_losses = []

# Stochastic Gradient Descent with Elastic Net Regularization (L1 and L2)
for epoch in range(num_epochs):
    # Shuffle the training data
    indices = np.random.permutation(len(X_train_b_SGD))
    X_train_b_shuffled = X_train_b_SGD[indices]
    y_train_shuffled = y_train_SGD[indices]

    # Mini-batch gradient descent
    for i in range(0, len(X_train_b_shuffled), batch_size):
        X_batch = X_train_b_shuffled[i:i + batch_size]
        y_batch = y_train_shuffled[i:i + batch_size]

        # Predict the output
        y_pred = X_batch @ theta

        # Compute the gradients with Elastic Net regularization (L1 and L2)
        gradients = -2 / len(X_batch) * X_batch.T @ (y_batch - y_pred) + alpha * (l1_ratio * np.sign(theta) + (1 - l1_ratio) * 2 * theta)

        # Update the weights
        theta -= learning_rate * gradients

    # Calculate Training RMSE after each epoch
    y_train_pred_SGD = X_train_b_SGD @ theta
    train_rmse_SGD = np.sqrt(np.mean((y_train_SGD - y_train_pred_SGD) ** 2))
    training_losses.append(train_rmse_SGD)

    # Calculate Validation RMSE after each epoch
    y_val_pred_SGD = X_val_b_SGD @ theta
    val_rmse_SGD = np.sqrt(np.mean((y_val_SGD - y_val_pred_SGD) ** 2))
    validation_losses.append(val_rmse_SGD)

    # Print the training and validation RMSE for this epoch
    print(f'Epoch {epoch + 1}/{num_epochs}, Train RMSE: {train_rmse_SGD:.4f}, Val RMSE: {val_rmse_SGD:.4f}, Diff: {train_rmse_SGD-val_rmse_SGD:.4f}')

# Plotting the RMSE for training and validation
mp.plot(training_losses, label='Train RMSE', color='blue', marker='o')
mp.plot(validation_losses, label='Validation RMSE', color='red', marker='x')
mp.xlabel('Epochs')
mp.ylabel('RMSE')
mp.title(f'Training and Validation RMSE (Elastic Net, alpha={alpha}, l1_ratio={l1_ratio})')
mp.legend()
mp.show()
```
